# South African Online Food Delivery Market Sizing Notebook

 #### 1. Problem Statement

This notebook estimates the market size of the South African online food delivery industry using the TAM, SAM, and SOM framework, based on population, smartphone penetration, and active delivery usage.

 #### 2. Key Assumptions

Population Segments
Total population: 65 million
Adults: 24 million
Smartphone users: 18 million
Regular food delivery users: 8 million
Market Behavior Assumptions
Average annual spend per active user: 
~2–3 orders per week

 #### 3. TAM (Total Addressable Market)

The total annual revenue potential if all active food delivery users participate fully in the market.

Formula
tam = potential_customers * avg_annual_spend

Calculation

TAM=8,000,000 × avg_annual_spend
Result

TAM = R112,996,752,000

 #### 4. Insight
Represents full national demand potential
Driven by urban middle-income digital consumers
Strong indicator of market attractiveness

 #### 5. SAM (Serviceable Available Market)

The portion of TAM that is realistically reachable given infrastructure and operational constraints.

Constraints:

- Urban residents only
- Smartphone owners
- Internet users
- Areas covered by delivery platforms
- 
Estimation Logic:

- Smartphone users: 18M
- Urban + connected + serviceable (~40-70%): ~6M users
- 
Formula
serviceable_customers = 3 500 000

sam = serviceable_customers * avg_annual_spend

 Result

SAM = RR49,436,079,000 (R49 Billion)

 Insight
This is the operational market
Concentrated in major cities:
Johannesburg
Cape Town
Durban
Pretoria
Success depends more on logistics coverage than demand

 #### 6. SOM (Serviceable Obtainable Market – Year 1)

The realistic share a new entrant can capture in the first year of operation.

Assumption: 

Conservative early-stage market share: 2% of SAM

Formula

SOM= SAM ×0.02
 Result

SOM = R988,721,580 (R988 Million)

 #### 7. Business Insights
7.1 Market structure
Small number of active users drive majority of revenue
Frequency of orders matters more than user base size
7.2 Competitive dynamics
Market dominated by logistics-heavy players
Key competitors rely on:
delivery speed
restaurant density
driver availability
7.3 Growth levers

To scale beyond SOM, focus on:

Increasing order frequency per user
Expanding restaurant partnerships
Reducing delivery time
Improving urban density coverage

  #### Conclusion

The South African food delivery market is:

Large (R112B TAM)
Highly concentrated (R49B SAM)
Competitive but accessible for niche entrants (R988M SOM)

Success depends less on demand creation and more on logistics efficiency and market positioning.
