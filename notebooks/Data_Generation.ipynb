```python
import numpy as np
import pandas as pd
import random
import os
```

```python
np.random.seed(42)
```

```python
n = 2000  # number of users

cities = ["Johannesburg", "Cape Town", "Durban", "Pretoria"]
income_levels = ["Low", "Medium", "High"]
apps = ["Uber Eats", "Mr D", "Bolt Food", "None"]

data = []

for i in range(n):
    age = np.random.randint(18, 60)
    
    if age < 25:
        income = np.random.choice(["Low", "Medium"], p=[0.7, 0.3])
    elif age < 35:
        income = np.random.choice(["Medium", "High"], p=[0.6, 0.4])
    else:
        income = np.random.choice(income_levels, p=[0.3, 0.5, 0.2])
    
    if income == "Low":
        orders = np.random.randint(0, 4)
        spend = np.random.randint(60, 120)
    elif income == "Medium":
        orders = np.random.randint(3, 11)
        spend = np.random.randint(100, 180)
    else:
        orders = np.random.randint(8, 21)
        spend = np.random.randint(150, 300)
    
    row = [
        i,
        age,
        random.choice(cities),
        income,
        random.choice(apps),
        orders,
        spend
    ]
    
    data.append(row)
```

```python
df = pd.DataFrame(data, columns=[
    "user_id", "age", "city", "income_level",
    "app_used", "orders_per_month", "avg_order_value"
])
```
