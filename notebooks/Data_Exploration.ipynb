```python
#analyze the distribution of the age variable
df['age'].hist(bins=30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
#How the distribution of the Income variable looks like
df['city'].hist(bins=30)
plt.xlabel('City')
plt.ylabel('Frequency')
plt.title('Distribution of City')
plt.show()
```

```python
#Pie chart to show the distribution of the income level variable
df['income_level'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title('Distribution of Income Level')
```

```python
#boxplot for app usage and orders per month
import seaborn as sns

# Create boxplot
plt.figure(figsize=(7, 5))
sns.boxplot(x="app_used", y="orders_per_month", data=df, palette="Set2", notch=True)

# Add title
plt.title("Distribution of Orders per Month by App Usage")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
